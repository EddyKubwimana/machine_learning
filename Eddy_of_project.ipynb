```python
numedata.reset_index(drop=True, inplace=True)
finalfactoData.reset_index(drop=True, inplace=True)
```

```python
data = pd.concat([numedata,finalfactoData], axis =1)
```

```python
len(data)
```

Splitting Data

```python
y= data["overall"]
```

```python
y
```

```python
x = data.drop(["overall"], axis = 1)
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(x,y,test_size = 0.8, random_state =42)
```

```python
alpha = 1.0
ridge = Ridge(alpha=alpha)
ridge.fit(Xtrain, Ytrain)

y_pred = ridge.predict(Xtest)
```

```python
mse = mean_squared_error(Ytest, y_pred)
r_squared = r2_score(Ytest, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r_squared)
```

```python
regression = LinearRegression()
```

```python
model = regression.fit(Xtrain,Ytrain)
```

```python
yprediction = model.predict(Xtest)
```

```python
mse2 = mean_squared_error(Ytest, yprediction)
r_squared2= r2_score(Ytest, yprediction)

print("Mean Squared Error:", mse2)
print("R-squared:", r_squared2)
```
